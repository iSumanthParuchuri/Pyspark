```
from pyspark.sql import *
tableDisease = spark.sql("select * from disease_csv")
tableHospitalData = spark.sql("select * from hospital_data_csv")
tablePatient = spark.sql("select * from patient_csv")
```

```
%sql
select * from disease_csv
```

```
from pyspark.sql.functions import monotonically_increasing_id
x = tableDisease.withColumn("Did", monotonically_increasing_id())
x = x.withColumn("DiseaseId", x.Did+1000)
tableDisease=x.select("Did","DiseaseId","Disease","Symptoms","Medicine/Manufacturer/Price","TreatmentN/Cost","TestName/Cost")
```

```
tableDisease.show()
```

```
from pyspark.sql.functions import col

#Disease data
storeDisease=tableDisease.select("Did","DiseaseId",col("Disease").alias("Disease Name"),"Symptoms")
storeDisease.show()
```

```
x = tableHospitalData.withColumn("Hid", monotonically_increasing_id())
x = x.withColumn("HospitalId", x.Hid+100)
tableHospitalData=x.select("Hid","HospitalId","Hospital_Name","Address","Hospital_ContactNo ","Hospital_Email_Id")
```

```
#Hospital data
storeHospital=tableHospitalData.select("Hid","HospitalId",col("Hospital_Name").alias("Hospital Name"),"Address",col("Hospital_ContactNo ").alias("Hospital Contact No"),col("Hospital_Email_Id").alias("Hospital Email Id"))
storeHospital.show()
```

```
x = tablePatient.withColumn("Pid", monotonically_increasing_id())
x = x.withColumn("PatientId", x.Pid+500)
print(x)
tablePatient=x.select("Pid","PatientId","Patient Name","gender","email","domain","DateOfbirth","Phone No","Address","Disease","Treatment Start Date","Treatment End Date","Hospital admitted","_c11")
```

```
#patient data
storePatient=tablePatient.select("Pid","PatientId","Patient Name","Gender","Email","Domain",col("DateOfBirth").alias("Date Of Birth"),"Phone No","Address")
storePatient.show()
```

```
#Patient Disease data
storePatientDiseaseTemp=tablePatient.select("PatientId",col("Disease").alias("Disease Name"),"Treatment Start Date", "Treatment End Date", "Hospital admitted")
merge=storePatientDiseaseTemp.join(storeDisease,on='Disease Name',how='inner')
storePatientDisease=merge.select("PatientId","DiseaseId","Treatment Start Date","Treatment End Date","Hospital admitted")
storePatientDisease.show()
```

```
from pyspark.sql.functions import explode
from pyspark.sql.functions import split
temp = tableDisease.select("Did","DiseaseId","Disease","Symptoms",explode(split("Medicine/Manufacturer/Price",",")).alias("x"), "TreatmentN/Cost", "TestName/Cost")
temp= temp.filter(temp.x != "").select("Did","DiseaseId","Disease","Symptoms",split("x",'/')[0].alias("Medicine"),split("x",'/')[1].alias("Manufacturer") ,split("x",'/')[2].alias("Price"),"TreatmentN/Cost","TestName/Cost")
temp = temp.select("Did","DiseaseId","Disease","Symptoms","Medicine","Manufacturer","Price",explode(split("TreatmentN/Cost",",")).alias("x"), "TestName/Cost")
temp= temp.filter(temp.x != "").select("Did","DiseaseId","Disease","Symptoms","Medicine","Manufacturer","Price",split("x",'/')[0].alias("TreatmentN"),split("x",'/')[1].alias("Cost"),"TestName/Cost")
temp = temp.select("Did","DiseaseId","Disease","Symptoms","Medicine","Manufacturer","Price","TreatmentN","Cost", explode(split("TestName/Cost",",")).alias("x"))
temp= temp.filter(temp.x != "").select("Did","DiseaseId","Disease","Symptoms","Medicine","Manufacturer","Price","TreatmentN","Cost", split("x",'/')[0].alias("TestName"),split("x",'/')[1].alias("TestCost"))
tableDiseaseMod=temp
```

```
#Medicine data
x=tableDiseaseMod.select("Medicine","Manufacturer","Price","DiseaseId").distinct().sort(col("DiseaseId").asc())
x = x.withColumn("Mid", monotonically_increasing_id())
x = x.withColumn("MedicineId", x.Mid+100)
storeMedicine=x.select("Mid","MedicineId","Medicine","Manufacturer","Price","DiseaseId")
storeMedicine.show()
```

```
#Test Data
temp=tableDiseaseMod.select("DiseaseId","TestName","TestCost").distinct().sort(col("TestName").asc())
x = temp.withColumn("Tid", monotonically_increasing_id())
x = x.withColumn("TestId", x.Tid+5000)
storeTest=x.select("Tid","TestId","TestName","TestCost")
storeTest.show()


#Test Disease Table
storeTestDisease=x.select("DiseaseId","TestId")
```

```
#Treatment Data
temp=tableDiseaseMod.select("TreatmentN","Cost").distinct().sort(col("TreatmentN").asc())
x = temp.withColumn("Trid", monotonically_increasing_id())
x = x.withColumn("TreatmentId", x.Trid+2000)
storeTreatment=x.select("Trid","TreatmentId","TreatmentN","Cost")
storeTreatment.show()
```
